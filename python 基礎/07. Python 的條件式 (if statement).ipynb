# Python 的 if 條件式

## 語法

```python
if (條件式)：
    欲執行的程式

```

````python
a = 30
b = 20

if a > b:
    print("a > b")
````

````python
a = 20
b = 20

if a > b:
    print("a > b")
elif a < b:
    print("a < b")
else:
    print("a = b")
````

````python
a = 20
b = 20

if a != b:
    print("a != b")
else:
    print("a = b")
````

````python
c = [1, 2, 3]

if c:
    for e in c:
        print(e)
````

````python
s = ""

if s:
    print("s 的值是：", s)
else:
    print("s 是空字串")
````

## 複合條件的使用

````python
a = 25

if 0 <= a <= 20:
    print("a 符合條件")
else:
    print("a 不符合條件")
````

````python
s = ""

if not s:
    print("s 是空字串")
````
