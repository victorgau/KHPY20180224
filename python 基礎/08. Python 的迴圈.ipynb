## for 迴圈

### 語法：

```python
for 變數 in 資料結構:
    欲執行之程式區塊
```

````python
# for 迴圈

A = range(10)

for i in A:
    print('{} \n'.format(i))
````

````python
help(range)
````

````python
# for 迴圈

A = range(1,10, 2)

for i, item in enumerate(A):    # enumerate: iterator
    print('{}: {} \n'.format(i, item))
````

## break 跟 continue

* break 用來跳出迴圈
* continue 用來繼續執行迴圈的下一輪迴

````python
for i in range(1, 10):
    print(i)
````

````python
for i in range(1, 10):
    if i % 5 == 0:
        break
    print(i)
````

````python
for i in range(1, 10):
    if i % 5 ==0:
        continue
    print(i)
````

## 巢狀迴圈

````python
for i in range(5):
    for j in range(5):
        print("i =", i, ", j =", j)
````

## while 迴圈

### 語法：

```python
while 條件式:
    欲執行之程式區塊
```

````python
# while 迴圈

count = 0
while (count < 8):
    print(count)
    count += 1
````

````python
# while 迴圈

count = 0
while (count < 8):
    count += 1
    if count % 2 == 0:
        continue
    print(count)
````

````python
# while 迴圈

count = 0
while (count < 8):
    count += 1
    if count % 2 == 0:
        break
    print(count)
````

````python
# while 迴圈

count = 0
while (count < 10):
    if count % 3 == 0:
        print(str(count))
    count += 1
````

````python
count = 0
while True:
    if count > 10:
        break
    count += 1
````
